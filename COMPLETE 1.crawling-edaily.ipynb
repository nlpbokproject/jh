```python
from bs4 import BeautifulSoup, Comment
from urllib.request import urlopen
import pandas as pd
from selenium import webdriver
import time
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
```

```python
import datetime, time

#이데일리 크롤링 코드

start_date = datetime.datetime.strptime("01-01-2017", "%d-%m-%Y") #시작 날짜
end_date = datetime.datetime.strptime("31-12-2017", "%d-%m-%Y") #끝 날짜

date_generated = [start_date + datetime.timedelta(days=x) for x in range(0, (end_date-start_date).days+1)]
date_list = []
#해당 날짜에 해당하는 뉴스 기사 URL 리스트
for date in date_generated:
    #print(date.strftime("%d-%m-%Y"))
    nowdate = str(date.year)+"."+str(date.month).zfill(2)+"."+str(date.day).zfill(2) # 현재 날짜 string 생성 ex.2017.05.21
    print(nowdate)
    date_list.append("https://search.naver.com/search.naver?where=news&query=%EA%B8%88%EB%A6%AC&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds="+nowdate+"&de="+nowdate+"&docid=&nso=so%3Ar%2Cp%3A%2Ca%3Aall&mynews=0&refresh_start=0&related=0")
    #네이버에서 '금리' 단어를 포함한 이데일리 뉴스 기사 검색
```

```python
len(date_list)
```

```python
df_1 = pd.read_csv('result.csv', names = ['title','date','url','script'])
```

```python
tmp =[]
driver = webdriver.Chrome("chromedriver.exe")
for date in date_list:
    driver.get(date)
    driver.set_page_load_timeout(30)
    driver.find_element_by_xpath("""//*[@id="news_popup"]/a""").click() #'네이버 뉴스' 클릭
    time.sleep(1) # 1초 쉼
    driver.find_element_by_xpath("""//*[@id="ca_1018"]""").click() #'이데일리' 항목 버튼 클릭
    time.sleep(1)
    driver.find_element_by_xpath("""//*[@id="_nx_option_media"]/div[2]/div[3]/button[1]/span""").click() #'확인' 버튼
    time.sleep(1) # 1초 쉼
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    a_tags = soup.find_all('a', class_='_sp_each_title') #해당 페이지의 모든 기사 제목
    print(a_tags[0].text)
    
    tmp.append(a_tags)
    
    while soup.find('a', class_='next') is not None: # '다음페이지' 버튼이 존재하면
        next_click = soup.find('a', class_='next') # '다음페이지' 버튼을 클릭
        url = 'https://search.naver.com/search.naver' + str(next_click['href'])
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        a_tags = soup.find_all('a', class_='_sp_each_title')
        tmp.append(a_tags)
        time.sleep(1)
        print(a_tags[0].text)
```

```python
date=''
prepro=''
cnt= 1
for j in range(len(tmp)):
    for i in tmp[j]:
         #제목
            print("제목 : " + i.get_text())
            title = i.get_text()
            #URL
            url = str(i['href'])
            response = requests.get(url)
            soup = BeautifulSoup(response.text,'html.parser')
            #print("URL : " + url)
            #날짜
            if soup.find('div', class_='dates') is not None:
                date = soup.find('div', class_='dates').find('p') # 해당 기사 URL 페이지에서 뉴스 날짜 부분 추출
                date = str(date)
                date = date[6:16]
                print(date)
            #print("date : "+ date)
            #본문
            if soup.find('div', class_='news_body') is not None:
                div_tag = soup.find('div', class_='news_body') # 해당 기사 URL 페이지에서 뉴스 기사 부분 추출
                prepro = str(div_tag)
                prepro = re.sub('\n','',prepro,0).strip()
                prepro = re.sub('<!--.+?-->','',prepro,0).strip()
                prepro = re.sub('<.+?>','',prepro,0).strip()
            #print(prepro)
            #print('',end='\n\n')
            new_df = pd.DataFrame({'title':title, 'date':date, 'url':url, 'script':prepro}, index=[0])
            # 새로운 데이터들을 데이터프레임에 저장
            df_1 = pd.concat([df_1, new_df]) # 기존 데이터프레임에 붙임
    df_1.to_csv("result.csv", encoding='UTF8')
```
