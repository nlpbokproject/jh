```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import re
```

```python
df = pd.DataFrame(columns=['idx', 'date'])
report = pd.read_csv("../final_data/bond_pre.csv")

for i in range(report.shape[0]):
    report_date = report["date"][i]
    report_date = report_date[0:7]
    df = df.append(pd.DataFrame([[i, report_date]], columns=['idx', 'date']), ignore_index=True)
    df = df.sort_values(["date"], ascending=True )
    
df.set_index('idx', inplace=True)
df = df.groupby("date").size()
print(df)
    
#     report_date = report_date.sort_values(["date"], ascending=[False])
    
```
