```python
from ekonlpy.sentiment import MPKO
from collections import defaultdict
def determiner_s(sentence, hawki_ls, dovis_ls):
    #문장의 톤을 구하는 함수. (문장 하나가 이 함수의 첫번째 변수로 들어옴.)
    hawki_cnt = 0
    dovis_cnt = 0
    mpko = MPKO(kind=1)
    tokens = mpko.tokenize(sentence)
    #문장별로 토큰화
    #print(tokens)
    for token in tokens:
        if token in hawki_ls.values:
            hawki_cnt += 1
            print(token+" hawk")
            #토큰이 매파에 해당하면 매파 개수 카운트 +1
        if token in dovis_ls.values:
            dovis_cnt += 1
            print(token+" dovis")
            #토큰이 비둘기파에 해당하면 비둘기파 개수 카운트 +1
    if hawki_cnt > dovis_cnt:
        return 1
        #매파 토큰이 더 많은 경우.
    elif hawki_cnt < dovis_cnt:
        return 0
        #비둘기파 토큰이 더 많은 경우.
    else:
        return -1
        #매파, 비둘키파 토큰의 수가 각각 같은 경우.

def determiner_p(p, hawki_ls, dovis_ls):
    tmp = p.split('.')
    # 마침표(.)을 단위로 문장으로 나눔
    hawki_cnt = 0
    dovis_cnt = 0
    for sentence in tmp:
        print(sentence)
        det = determiner_s(sentence, hawki_ls, dovis_ls)
        #나눠진 문장을 문장톤을 구하는 함수(본 함수의 위 함수)에 변수로 넣어 실행
        if det == 1:
            hawki_cnt +=1
            #매파 문장으로 판별 경우, hawk_cnt +1
        elif det == 0:
            dovis_cnt += 1
            #비둘기파 문장으로 판별 경우, dovis_cnt +1
        else:
            continue
    print('hawki : ' + str(hawki_cnt))
    print('dovis : ' + str(dovis_cnt))
    print("\nsentence end")
    try:
        tone = (hawki_cnt - dovis_cnt)/(hawki_cnt + dovis_cnt)
        #문서 톤 구하기
        return tone
    except:
        print("tone error")
        return None
    
#################################################################
#                        키워드 추출 함수                       #
#################################################################
def counting(p): # 단어 빈도수 계산 함수
    tmp_ls = []
    if type(p)==float:
        tmp_ls.append('none data') # 데이터가 없을 경우
    else:
        tmp = p.split('.')
        dic = defaultdict(lambda: 0)
        for token in tmp:
            #mpko = MPKO(kind=1)
            #tokens = mpko.tokenize(sentence)
            #for token in tokens:
            tok_l = token.split('/')
            if(len(tok_l[0])>=2):
                dic[token] += 1
        sort_dic = sorted(dic.items(), key=lambda x: x[1], reverse=True)
        
        for i in range(5): #키워드 추출 개수를 늘리시려면 range안의 숫자를 늘리시면 돼요
            tmp_ls.append(sort_dic[i]) 
    return tmp_ls
```

```python
#################################################################
#                        키워드 추출 코드                       #
#################################################################
i=1
#df['words']=''
import numpy as np
script_ls = df['ngram+unigram']
freq_ls = []
for p in script_ls:
    if p=='NaN':
        freq = 'nothing'
        freq_ls.append(freq)
        print(freq)
        #df[i]['words']=freq
        i += 1
        print(i)
    else:
        freq = counting(p)
        freq_ls.append(freq)
        print(freq)
        #df[i]['words']=freq
        i += 1
        print(i)
```
