```python
from glob import glob
import pandas as pd

#ngram과 unigram을 한 칼럼으로 합치는 코드.

#files = glob("./DATA/news_Yeonhap_info/processed/*.csv")
files = glob("./DATA/news_2006-2012/*.csv")
base_df = pd.read_csv("./DATA/rate.csv") # 콜금리 데이터를 불러옴

for file in files:
    df = pd.read_csv(file,lineterminator='\n')
    
    try:
        df.fillna('')
        df['ngram+unigram'] = ""
        df['ngram+unigram'] = df['tokens'] + ',' + df['ngram']
        #'ngram+unigram' 칼럼을 추가하여 token 칼럼의 데이터와 ngram 칼럼의 데이터를 연결
        print("success")
        
    except Exception as ex :
            print("fail")
            
    final_df = pd.merge(base_df, df, how='inner', on='date') # date를 기준으로 연결
    final_df[['date', 'ngram+unigram', 'label']].to_csv('./DATA/NBC/NBC-leftovers_step1.csv', mode='a', index=False)
    # 세 컬럼만 저장. 기존 파일에 append
```
