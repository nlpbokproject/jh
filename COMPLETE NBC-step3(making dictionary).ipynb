```python
#190528 edited - jh
dic = defaultdict(lambda : [0, 0])
dic['hawki']=[]
dic['dovis']=[]
for i in range(len(df)):
    label = df.loc[i]['label']
    if int(label)== 1: #label이 1이면
        dic['hawki'].append(df.loc[i]['ngrams']) #['hawki']에 추가
    else:
        dic['dovis'].append(df.loc[i]['ngrams']) #['dovis']에 추가
    if i % 1000 == 0:
        print(i)
```

```python
print(len(dic['hawki'])) #hawki 단어의 개수
print(len(dic['dovis'])) #dovis 단어의 개수
```

```python
for i in range(len(dic['hawki'])-len(dic['dovis'])):
    dic['dovis'].append('Nan')
#dovis 단어의 개수가 더 적으므로 리스트의 빈 부분에 Nan으로 채워줌
```
